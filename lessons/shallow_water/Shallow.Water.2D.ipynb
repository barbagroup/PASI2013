```python
from clawpack import riemann
from clawpack import pyclaw
import numpy
```

##Radial Dam

Our first foray into 2D shallow water simulations using PyClaw will be simulating the bursting of a radial dam.  

Imagine a column of water of some height, $h$, which is greater than the surrounding water level.  The dam holding the column in place is instantaneously removed and the water rushes out as a radial wave.  

```python
def set_initial_conditions(state, height_inside, x_velocity_inside, y_velocity_inside, height_outside, x_velocity_outside,y_velocity_outside, dam_radius, IC):
    ##Our domain is going to be 10x10, from -5 to 5 in both x and y.  We'll set 
    ##the origin of the dam at the point (-4, -4)
    x0 = -4.
    y0 = -4.
    xCenter = state.grid.x.centers
    yCenter = state.grid.y.centers
    
    Y,X = numpy.meshgrid(yCenter,xCenter)
    
    rad = numpy.sqrt((X-x0)**2 + (Y-y0)**2)
    
    if IC==1:
        state.q[0,:,:] = height_inside*(rad<=dam_radius) +\
                    height_outside*(rad>dam_radius)
        #state.q[1,:,:] = height_inside*x_velocity_inside*(rad<=dam_radius) + height_outside*x_velocity_outside*(rad>dam_radius)
        #state.q[2,:,:] = height_inside*y_velocity_inside*(rad<=dam_radius) + height_outside*y_velocity_outside*(rad>dam_radius)
    
    elif IC==2:
        eps=0.5
        state.q[0,:,:] = (height_inside + eps*numpy.exp(-(rad)**2/0.5))*\
                        (rad<=dam_radius)\
                        + height_outside*(rad>dam_radius)
    
    state.q[1,:,:] = height_inside*x_velocity_inside*(rad<=dam_radius) +\
                    height_outside*x_velocity_outside*(rad>dam_radius)
    state.q[2,:,:] = height_inside*y_velocity_inside*(rad<=dam_radius) +\
                    height_outside*y_velocity_outside*(rad>dam_radius)
    
```

```python
def setup(outdir='./_2d_output', IC=1):
    solver = pyclaw.ClawSolver2D(riemann.shallow_roe_with_efix_2D)
    solver.limiters = pyclaw.limiters.tvd.MC
    solver.dimensional_split=1 ##Enable Godunov splitting
    ###Add explanation of Godunov splitting###
    
    ##Apply Boundary Conditions##
    '''
    Here we're going to use wall boundary conditions at each boundary, so
    waves will be reflected back in to our domain.
    
    For 2D problems, PyClaw requires 4 boundary conditions, 2 each in the arrays
    bc_lower and bc_upper.  
    '''
    solver.bc_lower[0] = pyclaw.BC.wall
    solver.bc_lower[1] = pyclaw.BC.wall
    solver.bc_upper[0] = pyclaw.BC.wall
    solver.bc_upper[1] = pyclaw.BC.wall
    
    ###Define the domain geometry###
    
    ###Coordinates, note trailing '.' to make them floats###
    xlower = -5.
    xupper = 5.
    ylower = -5.
    yupper = 5.
    #################
    
    ###Number of Grid Points###
    nx = 150
    ny = 150
    ###########################
    
    ###Now use the pyclaw.Dimension function###
    ###This is similar to numpy.linspace###
    x = pyclaw.Dimension('x', xlower, xupper, nx)
    y = pyclaw.Dimension('y', ylower, yupper, ny)
    
    ###Using the dimensions defined, define the domain for PyClaw and the 
    ###state.  Note that the solver.num_eqn is defined by PyClaw for most basic
    ###simulations
    domain = pyclaw.Domain([x,y])
    state = pyclaw.State(domain, solver.num_eqn)
        
    grav = 1.0
    state.problem_data['grav'] = grav
    
    ###Initial Conditions###
    damRadius = 0.5
    
    ###Inside dam###
    height_inside = 2.
    x_velocity_inside = 0.
    y_velocity_inside = 0.
    
    ###Outside dam###
    height_outside = 1.0
    x_velocity_outside = 0.
    y_velocity_outside = 0.
    
    
    set_initial_conditions(state, height_inside, x_velocity_inside,\
                           y_velocity_inside, height_outside, x_velocity_outside,\
                           y_velocity_outside, damRadius, IC)
    
    ###Create a pyclaw.Controller###
    '''The Controller helps manage the simulation.  Once the controller is instantiated
    we can use it to define some parameters of the problem.  In the case below, the 
    simulation time is set to 10 seconds, the number of timesteps to log data for 
    is set to 50.  Then the problem we wish to solve is assigned
    and the data output directory is specified (as defined by our variable
    outdir)'''
    
    claw = pyclaw.Controller()
    claw.tfinal = 10
    claw.solution = pyclaw.Solution(state,domain)
    claw.solver = solver
    claw.outdir = outdir    
    claw.output_style=1
    
    claw.num_output_times = 50
    
    return claw
```

```python
outdir = './_2d_output'
claw = setup()
status = claw.run()
```

PyClaw has solved the problem according to the conditions we've supplied and has written a data file to our output directory for each solution step (in this case, we have 50 files).  

```python
import glob
import os
import numpy
import time
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

video_dir = '../videos'
```

```python
def pyclaw_2d_plotter(outdir):
    mx, my, xlow, xhigh, ylow, yhigh = import_sim_parameters(outdir)
    
    print "Importing fortran data files..."
    time1 = time.time()
    water_height, xmomentum, ymomentum = import_qfiles(mx, my, outdir)
    time2 = time.time()
    runtime = time2-time1
    print "Completed data import in %.2f seconds" % (runtime)
    
    time1 = time.time()
    print "Generating video frames"
    plot_water_surface(outdir, water_height, xmomentum, ymomentum, mx, my, xlow, xhigh, ylow, yhigh)
    time2 = time.time()
    runtime = time2-time1
    print "Finished generating plots in %.2f seconds" % (runtime)
```

```python
def plot_water_surface(outdir, plotdata, xmomentum, ymomentum, mx, my, xlow, xhigh, ylow, yhigh):
    
    x = numpy.linspace(xlow, xhigh, mx)
    y = numpy.linspace(ylow, yhigh, my)

    X, Y = numpy.meshgrid(x,y)
    for i in range(len(plotdata[:,0,0])):
        fig = plt.figure(figsize=(11,8))
        ax = fig.gca(projection='3d')
        #ax = plt.subplot2grid((2,3), (0,0), rowspan=2, colspan=2, projection='3d')
        surf = ax.plot_surface(X,Y,plotdata[i,:,:], rstride=1, cstride=1,\
                               cmap=cm.Blues, antialiased=True, linewidth=0)
        ax.set_zlim(0,2.)
        

        #ax1 = plt.subplot2grid((2,3), (0,2))
        #ax2 = plt.subplot2grid((2,3), (1,2))

        #ax1.plot(x,xmomentum[i,mx/2,:])
        #ax2.plot(y,ymomentum[i,my/2,:])
        #ax1.set_ylim(-0.6,0.6)
        #ax2.set_ylim(-0.05, 0.05)

        plt.savefig(video_dir + '/surf%03d.png' % i)
        plt.close()
```

```python
def import_qfiles(mx, my, outdir):
    '''Imports water height, x momentum and ymomentum data from ClawPack Fortran data 
    and stores them in numpy arrays'''

    filelist = sorted(glob.glob(outdir+'/*.q*'))

    water_height = numpy.zeros((len(filelist), mx, my))
    xmomentum = numpy.zeros((len(filelist), mx, my))
    ymomentum = numpy.zeros((len(filelist), mx, my))
    
    for i in range(len(filelist)):
        infile = numpy.genfromtxt(filelist[i], skip_header=9)
        water_height[i,:,:] = infile[:,0].reshape((mx,my))
        xmomentum[i,:,:] = infile[:,1].reshape((mx,my))
        ymomentum[i,:,:] = infile[:,2].reshape((mx,my))

    return water_height, xmomentum, ymomentum
```

```python
def import_sim_parameters(outdir):
    '''Imports simulation parameters from first (of many) ClawPack data files.
    Determines domain coordinates, grid size and step size and returns each'''
    with open(outdir+"/fort.q0000") as myfile:
        header = [myfile.next() for x in xrange(9)]

    mx = int(header[2].strip()[:header[2].strip().find(' ')])
    my = int(header[3].strip()[:header[3].strip().find(' ')])
    xlow = float(header[4].strip()[:header[4].strip().find(' ')])
    ylow = float(header[5].strip()[:header[5].strip().find(' ')])
    dx = float(header[6].strip()[:header[6].strip().find(' ')])
    dy = float(header[7].strip()[:header[7].strip().find(' ')])

    xhigh = numpy.round(xlow+mx*dx, decimals=1)
    yhigh = numpy.round(ylow+my*dy, decimals=1)


    return mx, my, xlow, xhigh, ylow, yhigh
```

##Periodic Boundary Conditions

Below is the same simulation as above, but with the four domain boundaries changed to periodic boundaries instead of stiff walls.  The only change to the code is to specify the different boundary conditions in the setup function.  

```python
def setup(outdir='./_2d_output'):
    solver = pyclaw.ClawSolver2D(riemann.shallow_roe_with_efix_2D)
    solver.limiters = pyclaw.limiters.tvd.MC
    solver.dimensional_split=1 ##Enable Godunov splitting
    ###Add explanation of Godunov splitting###
    
    ##Apply Boundary Conditions##
    '''
    Here we're going to use wall boundary conditions at each boundary, so
    waves will be reflected back in to our domain.
    
    For 2D problems, PyClaw requires 4 boundary conditions, 2 each in the arrays
    bc_lower and bc_upper.  
    '''
    solver.bc_lower[0] = pyclaw.BC.periodic
    solver.bc_lower[1] = pyclaw.BC.periodic
    solver.bc_upper[0] = pyclaw.BC.periodic
    solver.bc_upper[1] = pyclaw.BC.periodic
    
    ###Define the domain geometry###
    
    ###Coordinates, note trailing '.' to make them floats###
    xlower = -5.
    xupper = 5.
    ylower = -5.
    yupper = 5.
    #################
    
    ###Number of Grid Points###
    nx = 150
    ny = 150
    ###########################
    
    ###Now use the pyclaw.Dimension function###
    ###This is similar to numpy.linspace###
    x = pyclaw.Dimension('x', xlower, xupper, nx)
    y = pyclaw.Dimension('y', ylower, yupper, ny)
    
    domain = pyclaw.Domain([x,y])
    state = pyclaw.State(domain, solver.num_eqn)
    
    grav = 1.0
    state.problem_data['grav'] = grav
    
    
    ###Initial Conditions###
    damRadius = 0.5
    
    ###Inside dam###
    height_inside = 2.
    x_velocity_inside = 0.
    y_velocity_inside = 0.
    
    ###Outside dam###
    height_outside = 1.0
    x_velocity_outside = 0.
    y_velocity_outside = 0.
    
    
    set_initial_conditions(state, height_inside, x_velocity_inside,y_velocity_inside,\
                           height_outside, x_velocity_outside, y_velocity_outside, damRadius)
    
    claw = pyclaw.Controller()
    claw.tfinal = 10
    claw.solution = pyclaw.Solution(state,domain)
    claw.solver = solver
    claw.outdir = outdir
    
    claw.num_output_times = 50
    
    return claw
```

```python
outdir = './_2d_output_periodic'
claw = setup(outdir)
status = claw.run()
```
